### Merging the Data and Storing in a Database/Visualizing Data

```python
df_csv = pd.read_csv('CDC_data_clean.csv')  # Read in saved flat file data from Milestone 2.
df_web = pd.read_csv('website_data.csv')  # Read in saved website data from Milestone 3.
df_api = pd.read_csv('yelp_data.csv')  # Read in saved API data from Milestone 4.
```

```python
con = sqlite3.connect("CombinedData.db")  # Create database.
# Load individual dataframes into the database.
df_csv.to_sql("CSVTable", con)
df_web.to_sql("WebsiteTable", con)
df_api.to_sql("APITable", con)
```

```python
"""
Join the three tables on state name.

CSVTable.State_Name
WebsiteTable.State
APITable.State
"""

cur = con.cursor()
cur.execute("CREATE TABLE comb AS SELECT * FROM CSVTable LEFT JOIN WebsiteTable on CSVTable.State_Name = WebsiteTable.State \
            LEFT JOIN APITable on CSVTable.State_Name = APITable.State");
```

```python
df_comb = pd.read_sql_query("SELECT * FROM comb", con)  # Get combined database table to read into csv.
con.close()  # Close database.
#df_comb.to_csv('CombinedData.csv', index=False)
df_comb
```
